```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import io
df = pd.read_csv('Dataset_Masterclass_-_production_alimentaire4.csv', sep=';', encoding= 'latin-1')
print(df.head())
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
df= df.sort_values('emissions totales', ascending = False)
z= sns.barplot(data= df, x= 'Produit alimentaire', y='emissions totales')
z.set_title('Comparatif émissions totales')
plt.xticks(rotation=90)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
df= df.sort_values('emissions de gaz  effet de serre par 1000kcal (kgCO?eq par 1000kcal)', ascending= False)
d= sns.barplot(data= df, x= 'Produit alimentaire', y='emissions de gaz  effet de serre par 1000kcal (kgCO?eq par 1000kcal)') 
d.set_title('Comparatif emissions de gaz à effet de serre par 1000kcal (kgCO?eq par 1000kcal)')
plt.xticks(rotation=90)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
df= df.sort_values('emissions eutrophisation par 1000kcal (gPO?eq par 1000kcal)', ascending= False)
c= sns.barplot(data= df, x= 'Produit alimentaire', y='emissions eutrophisation par 1000kcal (gPO?eq par 1000kcal)') 
c.set_title('Comparatif emissions eutrophisation par 1000kcal (gPO?eq par 1000kcal)')
plt.xticks(rotation=90)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
df= df.sort_values('Prelevements eau douce par kilogramme (litres par kilogramme)', ascending= False)
b= sns.barplot(data= df, x= 'Produit alimentaire', y='Prelevements eau douce par kilogramme (litres par kilogramme)')
b.set_title('Comparatif prélevements eau douce par kilogramme (litres par kilogramme)')
plt.xticks(rotation=90)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
df= df.sort_values('Utilisation des sols par kilogramme (m² par kilogramme)', ascending=False)
a= sns.barplot(data= df, x= 'Produit alimentaire', y='Utilisation des sols par kilogramme (m² par kilogramme)')
a.set_title('Comparatif utilisation des sols par kilogramme (m² par kilogramme)')
plt.xticks(rotation=90)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
df= df.sort_values('Changement affectation des terres', ascending= False)
g= sns.barplot(data= df, x= 'Produit alimentaire', y='Changement affectation des terres')
plt.xticks(rotation=90)
g.set_title('Comparatif Changement affectation des terres par produit alimentaire')
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
df= df.sort_values('Consommation eau par kilogramme, ponderee par la rarete (litres par kilogramme)', ascending= False)
p= sns.barplot(data= df, x= 'Produit alimentaire', y='Consommation eau par kilogramme, ponderee par la rarete (litres par kilogramme)')
plt.xticks(rotation=90)
p.set_title('Comparatif Consommation eau par kilogramme par produit alimentaire')
```
